```python
# Filtering the features with 0 variance
no_var = df.var() == 0
print(df.var()[no_var])
print('\n')

# Drop features with 0 variance --- these features will not help in the performance of the model
df.drop(df.var()[no_var].index, axis = 1, inplace = True)
print(df.var())
print(df.shape)
```

```python
# Replace 0.526589 with 1 in the FLAG column and make its datatype bool
df['FLAG'] = df['FLAG'].replace(0.526589, 1)
df['FLAG'] = df['FLAG'].astype(int)
```

```python
# Features with high skewness
skewed_features = [
    'Avg min between sent tnx', 'Avg min between received tnx',
    'Time Diff between first and last (Mins)', 'Sent tnx', 'Received Tnx', 'Number of Created Contracts',
    'Unique Received From Addresses', 'Unique Sent To Addresses', 'min value received', 'max value received',
    'avg val received', 'min val sent', 'max val sent', 'avg val sent',
    'total transactions (including tnx to create contract', 'total Ether sent', 'total ether received',
    'total ether sent contracts', 'total ether balance', 'Total ERC20 tnxs', 'ERC20 total Ether received',
    'ERC20 total ether sent', 'ERC20 total Ether sent contract', 'ERC20 uniq sent addr', 'ERC20 uniq rec addr',
    'ERC20 uniq sent addr.1', 'ERC20 uniq rec contract addr', 'ERC20 min val rec', 'ERC20 max val rec',
    'ERC20 avg val rec', 'ERC20 min val sent', 'ERC20 max val sent', 'ERC20 avg val sent',
    'ERC20 uniq sent token name', 'ERC20 uniq rec token name'
]

# Apply logarithmic transformation to skewed features
for feature in skewed_features:
    # Apply log1p only to positive values
    df[feature] = np.where(df[feature] > 0, np.log1p(df[feature]), df[feature])

# Check the skewness again to verify if it has been reduced
skewness_after = df.skew()
print(skewness_after)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
X = df.drop(columns=['FLAG'])  # Features
y = df['FLAG']  # Target variable
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 123)
```

```python
sc = StandardScaler()
sc_train = sc.fit_transform(X_train)
```

```python
sc_df = pd.DataFrame(sc_train, columns=X_train.columns)
sc_df
```

# balancing using smote

```python
from imblearn.over_sampling import SMOTE
```

```python
oversample = SMOTE()
print(f'Shape of the training before SMOTE: {sc_train.shape, y_train.shape}')
```

```python
x_tr_resample, y_tr_resample = oversample.fit_resample(sc_train, y_train)
print(f'Shape of the training after SMOTE: {x_tr_resample.shape, y_tr_resample.shape}')
```

```python
# Target distribution before SMOTE
non_fraud = 0
fraud = 0

for i in y_train:
    if i == 0:
        non_fraud +=1
    else:
        fraud +=1

# Target distribution after SMOTE
no = 0
yes = 1

for j in y_tr_resample:
    if j == 0:
        no +=1
    else:
        yes +=1


print(f'BEFORE OVERSAMPLING \n \tNon-frauds: {non_fraud} \n \tFauds: {fraud}')
print(f'AFTER OVERSAMPLING \n \tNon-frauds: {no} \n \tFauds: {yes}')
```

### Logistic regression

```python
lr = LogisticRegression(random_state=42)
lr.fit(x_tr_resample, y_tr_resample)

# Transform test features
sc_test = sc.transform(X_test)
sc_preds = lr.predict(sc_test)
```

```python
print(classification_report(y_test, sc_preds))
```

```python
print(confusion_matrix(y_test, sc_preds))
```

### decision tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Create the decision tree classifier
dtree = DecisionTreeClassifier(random_state=42)  # Set random state for reproducibility

# Train the decision tree on the scaled training data
dtree.fit(x_tr_resample, y_tr_resample)
sc_test = sc.transform(X_test)
# Make predictions on the scaled testing data
sc_preds = dtree.predict(sc_test)

# Evaluate the decision tree's performance
print(classification_report(y_test, sc_preds))
```

```python
print(confusion_matrix(y_test, sc_preds))
```

### Random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_classifier = RandomForestClassifier(random_state=0)
```

```python
rf_classifier.fit(x_tr_resample, y_tr_resample)
sc_test = sc.transform(X_test)
```

# Balancing using class weights

```python
lr = LogisticRegression(class_weight='balanced',random_state=100)
lr.fit(sc_train, y_train)

# Transform test features
sc_test = sc.transform(X_test)

sc_preds = lr.predict(sc_test)
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Create the decision tree classifier
dtree = DecisionTreeClassifier(random_state=42,class_weight='balanced')  # Set random state for reproducibility

# Train the decision tree on the scaled training data
dtree.fit(sc_train, y_train)
sc_test = sc.transform(X_test)
# Make predictions on the scaled testing data
sc_preds = dtree.predict(sc_test)

# Evaluate the decision tree's performance
print(classification_report(y_test, sc_preds))
print(confusion_matrix(y_test, sc_preds))
```
